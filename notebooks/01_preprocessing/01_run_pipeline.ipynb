# LIMA Experiment Preprocessing Pipeline

This notebook runs the complete preprocessing pipeline for the LIMA experiment data.

## Pipeline Stages

1. **Raw Data Processing**: Convert JSON files to DataFrames
2. **Trial Data Processing**: Align effort data, calculate metrics
3. **Subjective Reports Processing**: Process anxiety/confidence ratings
4. **Mental Health Processing**: Score questionnaires (item-level and summed)
5. **Data Filtering**: Apply quality control, merge datasets
6. **Final Output**: Calculate effort metrics, create analysis-ready CSVs

## Outputs

- CSVs at every stage for transparency
- Mental health data: both item-level and summed scores
- Comprehensive logging and reports

### Stage 6: Vigor Computation

Computes per-trial vigor, per-epoch metrics, and condition cell means from `behavior_rich.csv`.
Can be run standalone if Stages 1-5 are already done.

```python
import sys
sys.path.insert(0, str(ROOT / 'scripts' / 'preprocessing'))
from compute_vigor import process_trial, compute_cell_means
from pathlib import Path
import pandas as pd

# Find latest stage2 and stage5 dirs
STAGE2_DIR = sorted(Path(OUTPUT_BASE_DIR).glob('stage2_*'))[-1]
STAGE5_DIR = sorted(Path(OUTPUT_BASE_DIR).glob('stage5_*'))[-1]
VIGOR_DIR = ROOT / 'results' / 'stats' / 'vigor_analysis'
VIGOR_DIR.mkdir(parents=True, exist_ok=True)
EXCLUDE = [154, 197, 208]

# Load from pickle (preserves list columns)
print(f'Loading {STAGE2_DIR / "processed_trials.pkl"}...')
td = pd.read_pickle(STAGE2_DIR / 'processed_trials.pkl')

# Add subj mapping
mapping = pd.read_csv(STAGE5_DIR / 'subject_mapping.csv')
pid_to_subj = dict(zip(mapping['participantID'], mapping['subj']))
td['subj'] = td['participantID'].map(pid_to_subj)
td = td[td['subj'].notna()]
td['subj'] = td['subj'].astype(int)
td = td[~td['subj'].isin(EXCLUDE)]
n_subj = td['subj'].nunique()
print(f'Loaded {len(td)} trials, {n_subj} subjects')

# Process all trials
print('Computing vigor metrics...')
all_epochs = []
for _, row in td.iterrows():
    all_epochs.extend(process_trial(row))
epoch_df = pd.DataFrame(all_epochs)

# Save
trial_vigor = epoch_df[epoch_df['epoch'] == 'full'].copy()
trial_vigor.to_csv(STAGE5_DIR / 'trial_vigor.csv', index=False)
print(f'  trial_vigor: {len(trial_vigor)} trials')

epoch_df.to_csv(VIGOR_DIR / 'vigor_metrics.csv', index=False)
print(f'  vigor_metrics: {len(epoch_df)} rows')

cell_means = compute_cell_means(epoch_df)
cell_means.to_csv(VIGOR_DIR / 'cell_means.csv', index=False)
print(f'  cell_means: {len(cell_means)} cells')
print('\nStage 6 complete.')
```
