# LIMA Experiment Preprocessing Pipeline

This notebook runs the complete preprocessing pipeline for the LIMA experiment data.

## Pipeline Stages

1. **Raw Data Processing**: Convert JSON files to DataFrames

5. **Data Filtering**: Apply quality control, merge datasets
6. **Final Output**: Calculate effort metrics, create analysis-ready CSVs

## Outputs

- CSVs at every stage for transparency
- Mental health data: both item-level and summed scores
- Comprehensive logging and reports

Reshape and clean the mental health data.

```python
# NOTE: This cell requires external Prolific demographics file not included in repo.
# Update the paths below when the pipeline has been run and output files are available.
# The paths below reference a previous run's output directory (ProcessingPipelineData_FINAL)
# which is not included in the repo. Re-run stages 4 and 5 first to generate these files.

# Load data
# df_raw = pd.read_csv(ROOT / 'data' / 'exploratory_350' / 'processed' / '<stage4_dir>' / 'mental_health_items_wide.csv')
# subjs = pd.read_csv(ROOT / 'data' / 'exploratory_350' / 'processed' / '<stage5_dir>' / 'subject_mapping.csv')

# df_raw = pd.read_csv('/Users/nokada/Documents/CALTECH/LIMA_Analysis/Data/ProcessingPipelineData_FINAL/stage4_mental_health_20260102_104909/mental_health_items_wide.csv')
# subjs = pd.read_csv('/Users/nokada/Documents/CALTECH/LIMA_Analysis/Data/ProcessingPipelineData_FINAL/stage5_filtered_data_20260102_104909/subject_mapping.csv')

# df = df_raw[df_raw['participantID'].isin(subjs['participantID'])]

# print(f"Dataset shape: {df.shape}")
# print(f"Participants: {len(df)}")
# print(f"Columns: {len(df.columns)}")
# print("\nFirst few rows:")
# df.head()
```

```python
# Identify questionnaire columns
all_cols = [col for col in df.columns if col != 'participantID']

# Parse by questionnaire
item_info = {}
for col in all_cols:
    prefix = col.rsplit('_item_', 1)[0] if '_item_' in col else None
    if prefix:
        if prefix not in item_info:
            item_info[prefix] = []
        item_info[prefix].append(col)

print("Questionnaires detected:")
for prefix, cols in item_info.items():
    print(f"  {prefix}: {len(cols)} items")
```

```python
# EXCLUDE STICSA (NON-NEGOTIABLE)
# Identify STICSA columns
sticsa_cols = item_info.get('STICSA', [])

print(f"⚠️ DROPPING {len(sticsa_cols)} STICSA items from dataframe")

# Drop from dataframe
df = df.drop(columns=sticsa_cols, errors='ignore')

# Select items to include
included_prefixes = ['AMI', 'DASS21', 'MFIS', 'OASIS', 'PHQ9', 'STAI', 'STAI_2']
item_cols = []
for prefix in included_prefixes:
    if prefix in item_info:
        item_cols.extend(item_info[prefix])

print(f"\nTotal items for analysis: {len(item_cols)}")
print(f"\nItems per measure:")
for prefix in included_prefixes:
    if prefix in item_info:
        print(f"  {prefix}: {len(item_info[prefix])} items")
```

```python
# Check response ranges (should match specification)
print("Response ranges by questionnaire:")
print("=" * 50)

expected_ranges = {
    'AMI': (0, 4),
    'DASS21': (0, 3),
    'MFIS': (0, 4),
    'OASIS': (0, 4),
    'PHQ9': (0, 3),
    'STAI': (0, 3),
    'STAI_2': (0, 3)
}

for prefix in included_prefixes:
    if prefix in item_info:
        cols = item_info[prefix]
        min_val = df[cols].min().min()
        max_val = df[cols].max().max()
        expected = expected_ranges.get(prefix, 'Unknown')
        status = "✓" if (min_val, max_val) == expected else "⚠️"
        print(f"{prefix:10s}: observed ({min_val:.0f}-{max_val:.0f}), expected {expected} {status}")
```

```python
# Response scale maxima
SCALE_MAX = {
    'AMI': 4,
    'DASS21': 3,
    'MFIS': 4,
    'OASIS': 4,
    'PHQ9': 3,
    'STAI': 3,
    'STAI_2': 3
}

# STAI reverse items (0-indexed)
STAI_REVERSE_ITEMS = [0, 1, 4, 7, 9, 10, 14, 15, 18, 19]  # State anxiety positively-worded
STAI_2_REVERSE_ITEMS = [0, 2, 5, 6, 9, 12, 13, 15, 18]    # Trait anxiety positively-worded

# Apply reverse scoring
df_scored = df.copy()
reverse_log = []

# AMI: ALL items
ami_cols = [col for col in item_cols if col.startswith('AMI_')]
for col in ami_cols:
    df_scored[col] = SCALE_MAX['AMI'] - df_scored[col]
    reverse_log.append(col)
print(f"AMI: Reversed {len(ami_cols)} items")

# STAI State
stai_reversed = []
for idx in STAI_REVERSE_ITEMS:
    col = f'STAI_item_{idx}'
    if col in df_scored.columns:
        df_scored[col] = SCALE_MAX['STAI'] - df_scored[col]
        stai_reversed.append(col)
        reverse_log.append(col)
print(f"STAI State: Reversed {len(stai_reversed)} items")

# STAI Trait
stai2_reversed = []
for idx in STAI_2_REVERSE_ITEMS:
    col = f'STAI_2_item_{idx}'
    if col in df_scored.columns:
        df_scored[col] = SCALE_MAX['STAI_2'] - df_scored[col]
        stai2_reversed.append(col)
        reverse_log.append(col)
print(f"STAI Trait: Reversed {len(stai2_reversed)} items")

print(f"\nTotal items reverse-scored: {len(reverse_log)}")
print("\n✓ After reverse scoring: HIGHER values = MORE SEVERE symptoms")
```

```python
# Extract item data
item_df = df_scored[item_cols].copy()

# Per-item missingness
item_missing = item_df.isnull().sum()
item_missing_pct = (item_missing / len(item_df)) * 100

print("Missing data per questionnaire:")
print("=" * 50)
for q in included_prefixes:
    q_prefix = q if q != 'STAI_2' else 'STAI_2'
    if q == 'STAI':
        q_cols = [c for c in item_cols if c.startswith('STAI_item_')]
    else:
        q_cols = [c for c in item_cols if c.startswith(f'{q}_')]
    if q_cols:
        q_missing = item_df[q_cols].isnull().sum().sum()
        q_total = len(item_df) * len(q_cols)
        print(f"  {q:10s}: {q_missing:5d}/{q_total:6d} ({100*q_missing/q_total:5.2f}%)")
```

```python
# Per-participant missingness
participant_missing = item_df.isnull().sum(axis=1)
participant_missing_pct = participant_missing / len(item_cols)

# PRIMARY APPROACH: Listwise deletion with threshold
MISSING_THRESHOLD = 0.10

n_original = len(df_scored)
high_missing_mask = participant_missing_pct > MISSING_THRESHOLD
n_excluded = high_missing_mask.sum()

print("PRIMARY APPROACH: Listwise Deletion")
print("=" * 50)
print(f"Threshold: >{MISSING_THRESHOLD*100:.0f}% missing items per participant")
print(f"Participants excluded: {n_excluded} ({100*n_excluded/n_original:.1f}%)")

# Apply
df_clean = df_scored[~high_missing_mask].copy()
item_df_clean = df_clean[item_cols].dropna()
n_complete = len(item_df_clean)

print(f"Complete cases for EFA: {n_complete}")
```
